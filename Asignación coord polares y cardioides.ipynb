```python
import pandas as pd
import numpy as np
np.random.seed(123)

df1 = pd.DataFrame({
    'de': np.sort(np.random.normal(loc = 4, scale = 1, size=96)),
    'dl': np.sort(np.random.normal(loc=4.5, scale=1.2, size=96)),
    'ddd': np.repeat(prog(r=7, n=25, a1=15), 4)
})

df1['localidad'] = np.repeat(['l1','l2']*24, 2)
df1.head()
```

### Convirtiendo coordenadas polares

```python
import numpy as np

def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    theta = np.degrees(np.arctan2(y, x))
    return(rho, theta)

coordpde = cart2pol(df1['ddd'], df1['de'])
#print(coordpde)
df3 = pd.DataFrame({
    'rho':coordpde[0],
    'theta': coordpde[1],
    })
print(df3) 

coordpdl = cart2pol(df1['ddd'], df1['dl'])
#print(coordpdl)
df4 = pd.DataFrame({
    'rho':coordpdl[0],
    'theta':coordpdl[1],
    })
print(df4)


coordpdle = cart2pol(df1['dl'], df1['de'])
#print(coordpde)
df5 = pd.DataFrame({
    'rho':coordpdle[0],
    'theta': coordpdle[1],
    })
print(df3) 
```
